```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
#Defining the dataset

X = [i for i in range(-100,100)]
Y = [i for i in range(-100,100)]
Z =  [i - random.randint(0,100) for i in range(-100,100)]


X = [[X[i], Y[i]] for i in range(len(X))]

X_data, Y_data = np.array(X), np.array(Z)

```

```python
#Test Train split

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_data,Y_data,test_size=0.05)


m = X_train.shape
```

```python
#Defining weights and bias 
# Y = WX  +  B

W = np.random.rand(1, 1)
B = np.random.rand(1, 1)

print("W size :",W.shape)
print("B size :",B.shape)
print("X train size :",X_train.shape)
print("X test size :",X_test.shape)
print("Y train size :",Y_train.shape)
print("Y test size :",Y_test.shape)
print("W mat",W)
print("B mat",B)


# Y is an array of size (M,1) , X = (M,N) here N = 1, hence W = (1,1), B = (M,1)
```

```python
#Defining the learning rate

learning_rate = 0.0001
```

```python
# Implementing the linear regression Fully 
#Defining weights and bias 
# Y = WX  +  B
import random 

w0 = 0
w1 = 0
B = 0

#W = 9992.95668876
#B = -16632126.19550556


#Defining the learning rate

learning_rate = 0.0001

#Defining the number of epochs
epochs = 1000
W_list = []
B_list = []
error_tracker = []

m = X_train.shape[0]

for e in range(epochs):
    dj_dw1 = 0
    dj_db = 0
    dj_dw2 = 0
    error = 0
    #Implementing the forward Pass  Y_dash  =  W * X + B
    for (i,x) in enumerate(X_train):
     #   print(i,x,Y_train[i])
        y = w0 * x[0] + w1 * x[1] + B
        error = error + (y - Y_train[i])**2
        dj_dw1 = dj_dw1 +  (y - Y_train[i]) *x[0]
        dj_db = dj_db +  (y - Y_train[i])
        dj_dw2 = dj_dw2 +  (y - Y_train[i]) *x[1]
    #Gradient descentant
    error_tracker.append(error/m)
    w0 = w0 - learning_rate * (1/m) * dj_dw1
    w1 = w1 - learning_rate * (1/m) * dj_dw2
    B = B - learning_rate * (1/m) * dj_db
        
#Testing the model with test data

predicted_Y = []
for i,x in enumerate(X_train):
    y = w0 * x[0] + w1 * x[1] + B
    predicted_Y.append(y)
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')

# Plotting data (X, Y, Z)
ax.scatter(X_train[:, 0], X_train[:, 1], Y_train)

# Label the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

ax = fig.add_subplot(122, projection='3d')

# Plotting data (X, Y, Z)
ax.scatter(X_train[:, 0], X_train[:, 1], np.array(predicted_Y))

# Label the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')


# Show the plot
plt.show()
```
